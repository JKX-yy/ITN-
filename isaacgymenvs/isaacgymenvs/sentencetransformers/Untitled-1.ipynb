```python
from sentence_transformers import SentenceTransformer, util
# import sys
# print(sys.path)
model = SentenceTransformer("all-MiniLM-L6-v2")

source_task = [
    "There is a model hexagonal nut on the table, and the goal of Frank's robotic arm is to pick up the model hexagonal nut from the table.",
    "Franks arm grasps the model hexagonal nut and approaches the model hexagonal stud placed on the table and places the nut on the bolt.",
    "On the tabletop were placed nuts of type M3 and bolts of type M3, where the nuts were placed on top of the bolts, and Frank's arm was required to screw the nuts to the bottom of the bolts",
    "On the tabletop were placed nuts of type M18 and bolts of type 18, where the nuts were placed on top of the bolts, and Frank's arm was required to screw the nuts to the bottom of the bolts",
    "There is a model M6 nut on the table, and the goal of Frank's robotic arm is to pick up the model M6 nut from the table..",
    "Franks arm grasps the model M6 nut and approaches the model M6 stud placed on the table and places the nut on the bolt",
    "On the tabletop were placed nuts of type M6 and bolts of type M6, where the nuts were placed on top of the bolts, and Frank's arm was required to screw the nuts to the bottom of the bolts.",
    "On the tabletop were placed nuts of type M3 and bolts of type M3, where the nuts were placed on top of the bolts, and Frank's arm was required to screw the nuts to the bottom of the bolts.",
    "There is a model M18 nut on the table, and the goal of Frank's robotic arm is to pick up the model M18 nut from the table.",
    "There is a model zreeom nut on the table, and the goal of Frank's robotic arm is to pick up the model zreeom nut from the table.",
    "Franks arm grasps the model M18 nut and approaches the model M18 stud placed on the table and places the nut on the bolt.",
    "There is a model M3 nut on the table, and the goal of Frank's robotic arm is to pick up the model M3 nut from the table.",
]

target_task=["On the tabletop were placed nuts of type M3 and bolts of type M3, where the nuts were placed on top of the bolts, and Frank's arm was required to screw the nuts to the bottom of the bolts.",]


# Encode all sentences
embeddings1 = model.encode(source_task)
embeddings2 = model.encode(target_task)
# Compute cosine similarity between all pairs
cos_sim = util.cos_sim(embeddings2, embeddings1)
print(cos_sim.shape)
print(cos_sim)
```
